# Tic-Tac-Toe

```python
def grid_printer(cells):
    print("---------")
    print("|", cells[0], cells[1], cells[2], "|")
    print("|", cells[3], cells[4], cells[5], "|")
    print("|", cells[6], cells[7], cells[8], "|")
    print("---------")
    
def win_analyzer(cells):
    wincomb = [[0, 1, 2], [0, 3, 6], [0, 4, 8], [6, 7, 8], 
               [2, 5, 8], [3, 4, 5], [1, 4, 7], [2, 4, 6]]
    fin = ''
    for i,j,k in wincomb:
        if (cells[i] != ' ') and (cells[i] == cells[j]) and (cells[j] == cells[k]):
            if fin != '':
                fin = "Impossible"
            else:
                fin = fin = f"{cells[i]} wins"          
    if fin == '' and ' ' not in cells:
        fin = "Draw"
    elif fin == '' and " " in cells:
        fin = "Game not finished"  
    if cells.count("X") > cells.count("O") + 1 or cells.count("O") > cells.count("X") + 1:
        fin = "Impossible"
    return fin

def input_taker(cells, move):
    coord = input("Enter the coordinates: ")
    if len(coord) == 3 and coord[0].isnumeric() and coord[2].isnumeric():
        coord = [int(i)-1 for i in coord.split()]
        if coord[0] < 3 and coord[1] < 3:
            coord = 3*coord[0]+coord[1]
            if cells[coord] == " ":
                cells[coord] = move
            else:
                print("This cell is occupied! Choose another one!")
        else:
            print("Coordinates should be from 1 to 3!")
    else:
        print("You should enter numbers!")
    return cells
        
def move_switcher(move):
    if move == "O":
        return "X"
    else:
        return "O"
        
#Start the programm
cells = [' ' for _ in range(9)]
grid_printer(cells)

status = 'Game not finished'
move = "O"
while status == 'Game not finished':
    move = move_switcher(move)
    cells = input_taker(cells, move)
    grid_printer(cells)
    status = win_analyzer(cells)
print(status)
```
